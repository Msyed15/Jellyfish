```python
import matplotlib.pyplot as plt

# Get unique years
years = df_PP['Year'].unique()
num_years = len(years)

# Define the number of columns for the subplots
num_cols = 2

# Calculate the number of rows needed
num_rows = (num_years + num_cols - 1) // num_cols

# Create the subplots
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 6))
fig.tight_layout(pad=3.0)

# Iterate over the years and create pie charts
for i, year in enumerate(years):
    # Get the corresponding axes for the subplot
    row = i // num_cols
    col = i % num_cols
    ax = axs[row, col]

    # Prepare data for the current year
    labels = ['Palestine', 'Israel']
    sizes_PP = df_PP[df_PP['Year'] == year]['Value'].str.replace(',', '').astype(float)
    sizes_PI = df_PI[df_PI['Year'] == year]['Value'].str.replace(',', '').astype(float)

    # Combine data for both countries
    sizes = [sizes_PP.sum(), sizes_PI.sum()]

    # Plot the pie chart
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

    # Set title with the current year
    ax.set_title(f'Year {year}')

# Add a big title for all the pie charts
fig.suptitle('Poverty headcount ratio at $6.85 a day (2017 PPP) (% of population)', y=1.05)

# Display the chart
plt.show()



```
